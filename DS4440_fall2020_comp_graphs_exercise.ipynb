3. Let $x=3$, $w_1=5$, $w_2=2$, $b_2=3$, $b_1=1$ and $y$=40. Run through a **forward pass**, yielding $\mathcal{L}$

$\hat{y} = w_2(w_1 x + b_1) + b_2 = 2(5\cdot 3 + 1) + 3 = 35$

$\mathcal{L} = (y - \hat{y})^2 = (40 - 35)^2 = 25$

4. (*Foreshadowing next lecture*) What is $\frac{\partial \mathcal{L}}{\partial w_2}$? (Ignore other terms).

$\frac{\partial \mathcal{L}}{\partial w_2} = \frac{1}{n}\sum\frac{\partial}{\partial w_2} (y - \hat{y})^2$

$ = \frac{1}{n}\sum\frac{\partial}{\partial w_2} (y - w_2(w_1 x + b_1) - b_2)^2$

$ = \frac{1}{n}\sum 2(y - w_2(w_1 x + b_1) - b_2)\frac{\partial}{\partial w_2} (y - w_2(w_1 x + b_1) - b_2)$

$ = \frac{1}{n}\sum 2(y - w_2(w_1 x + b_1) - b_2)(w_1 x + b_1)$

$ = \frac{2}{n}\sum ((w_1 x + b_1)y - w_2(w_1 x + b_1)^2 - (w_1 x + b_1)b_2)$
